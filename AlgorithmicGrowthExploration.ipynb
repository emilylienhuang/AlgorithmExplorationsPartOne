**Deliverable 1**

Let

f(n) = 1/2 n(n-1) + 10

and

g(n) = n^2

First: plot with n ranging from 1 to 10

```python
import matplotlib.pyplot as plt
import numpy as np

#set x-axis and y-axis values
n = np.arange(1, 10, 0.1)
f = 0.5*(n*(n-1)) + 10
g = n * n


plt.plot(n,g,"purple", label="g(n)")
plt.plot(n,f, "green", label="f(n)")
plt.legend()
```

It is unclear in the first plot if g(n) and f(n) have the same order of growth as before n = 4, f(n) grows faster than g(n), and g(n) grows faster that f(n) when n is greater than 4. To be able to make a truly educated guess we need to see what happens as n gets even larger. Let's see what happens when we zoom out. Next, plot with n ranging from 1 to 10^6 ...

```python
#set x-axis and y-axis values
n = np.arange(1, pow(10,6), 0.1)
f = 0.5*(n*(n-1)) + 10
g = n * n


plt.plot(n,g,"purple", label="g(n)")
plt.plot(n,f,"green", label="f(n)")
plt.legend()
```

Here, we see that neither f(n) or g(n) shrink down to a flat line, so the orders of growth appear to "hold up" against each other. Let's try to state with definitive evidence using the limit theorem the relationships between f(n) and g(n) using orders of growth.

***Deliverable 2: The Limit Theorem ***


```python
#from 1 to 10
n = np.arange(1, 10, 0.1)
f = 0.5*(n*(n-1)) + 10
g = n * n

lim = f/g

plt.plot(n,lim, label="lim")
plt.legend()
```

This plot seems to convey that as n gets larger the limit approaches a constant that is greater than 0, so the order of growth would be the same. We don't know the whole story from this plot though, so we need to look further out to validate our assumptions. Let's look at what happens if we zoom out.


```python
#from 1 to 10^6
n = np.arange(1, pow(10,6), 0.1)
f = 0.5*(n*(n-1)) + 10
g = n * n
lim = f/g

plt.plot(n,lim, label="lim")
plt.legend()
```

Here, it is clear to see that the limit never drops to or even below 0. Rather, it is a positive constant. So f(n)  is in Theta(g(n)) which implies it also exists in O(g(n)) and Omega(g(n)). Therefore, the OOG of f(n) can be categorized by O(g(n)), Theta(g(n)), and Omega(g(n)) .

***Deliverable 3 ***

Now let's examine different orders of growth.

Let f(n) = sqrt(n^5 + 3n + 1)


and


g(n) = 5(n^2)


Repeat deliverables 1 and 2 for the given functions.

```python
# Deliverable 3.1: plot functions side by side first from 1 to 10
#then 1 to 10^6

n = np.arange(1, 10, 0.1)
f = pow((pow(n,5) + 3*n + 1), 0.5)
g = 5 *(n * n)

plt.plot(n,g,"red", label="g(n)")
plt.plot(n,f,"blue", label="f(n)")
plt.legend()
```

Tricky, tricky! From 1 to 10 they look like they may be the same order of growth. From only 1 to 10 isn't the whole story though! We need to see if there's a generalized statement we can make for an arbitrary large N. So when we zoom out...

```python
# Deliverable 3.1: plot functions side by side first from 1 to 10
#then 1 to 10^6
#set x-axis and y-axis values
n = np.arange(1, pow(10,6), 0.1)
f = pow((pow(n,5) + 3*n + 1), 0.5)
g = 5 *(n * n)


plt.plot(n,g,"red",label="g(n)")
plt.plot(n,f,"blue", label="f(n)")
plt.legend()
```

It is clear that f(n) grows WAY faster than g(n) since g(n) flattens out to near zero in respect to f(n) which keeps growing. Let's try to say something definitive about the relationships between f and g using the limit theorem...

```python
# Deliverable 3.2: The limit from 1 to 10 then 1 to 10^6
n = np.arange(1, 10, 0.1)
f = pow((pow(n,5) + 3*n + 1), 0.5)
g = 5 *(n * n)
lim = f/g

plt.plot(n,lim,"pink", label="lim")
plt.legend()
```

From 1 to 10, the limit looks like it is above 0, so it looks like they could be the same order of growth, but we need more information...

```python
# Deliverable 3.2: The limit from 1 to 10 then 1 to 10^6
import numpy as np
n = np.linspace(1, pow(10,6), 100)
f = pow((pow(n,5) + 3*n + 1), 0.5)
g = 5 *(n * n)
lim = f/g

plt.plot(n,lim,"pink", label="lim")
plt.legend()
```

Whoa! It looks like this limit is heading out towards infinity. Slowly at that, but nonetheless infinity. So f(n) seems to grow faster than g(n) given the limit test. Just to check this conclusion, let's zoom even further out...

```python
n = np.linspace(1, pow(10,50), 100)
#n = np.arange(1, pow(10,6), 0.1)
f = pow((pow(n,5) + 3*n + 1), 0.5)
g = 5 *(n * n)
lim = f/g

plt.plot(n,lim,"pink", label="lim")
plt.legend()
```

Here we can say definitively that the limit is going out to infinity slowly but surely. Therefore, f(n) is in Omega(g(n)), but not Theta(g(n)) or O(g(n)).

***Deliverable 4:***
Repeat deliverable 3 with the functions:

g(n) = sqrt(n)

```python
#Deliverable 4.1 Plotted side by side from 1 to 10 then 1 to 10^6
import math
n = np.arange(1, 10, 0.1)
f = np.log(n)
g = pow(n, 0.5)

plt.plot(n,g,"black", label="g(n)")
plt.plot(n,f,"yellow", label="f(n)")
plt.legend()
```

From 1 to 10, it's hard to say anything about the relationships between f and g. Zoomed out...

```python
#Deliverable 4.1 Plotted side by side from 1 to 10 then 1 to 10^6
import math
n = np.arange(1, pow(10,6), 0.1)
f = np.log(n)
g = pow(n, 0.5)

plt.plot(n,g,"black", label="g(n)")
plt.plot(n,f,"yellow", label="f(n)")
plt.legend()
```

It is apparent that g(n) has a much larger order of growth than f(n) because f(n) flattens out to zero in comparison.

```python
#Deliverable 4.2: The limit f(n)/g(n) from 1 to 10 then 1 to 10^6

import math
n = np.arange(1, 10, 0.1)
f = np.log(n)
g = pow(n, 0.5)
lim = f/g

plt.plot(n,lim,"brown", label="lim")
plt.legend()
```

From 1 to 10 it looks like they may be the same growth rate as that limit is above 0 from n = 1 to 10 od further appears to be leveling off and approaching some constant value, but we are unsure as n approaches infinity. Let's take a look from 1 to 10^6...

```python
import math
n = np.arange(1, pow(10,6), 0.1)
f = np.log(n)
g = pow(n, 0.5)
lim = f/g

plt.plot(n,lim,"orange", label="lim")
plt.ylim(0,0.05)
plt.legend()
```

Here, it is clear to see that the limit at infinity approaches a constant above 0, so it appears that it may be the same order of growth, but let's zoom even FURTHER out...

```python
n = np.linspace(1, pow(10,50), 100)
f = np.log(n)
g = pow(n, 0.5)
lim = f/g

plt.plot(n,lim,"orange", label="lim")
plt.ylim(0,0.05)
plt.legend()
```

Here, it is apparent that the limit goes to 0 which means that g(n) grows faster than f(n); accordingly, f(n) is in O(g(n)) and not Theta(g(n)) or Omega(g(n)).

***Deliverable 5:***
Repeat the same deliverables with the functions log_2(n) and log_10(n).

```python
#Deliverable 5.1: Graph the functions from 1 to 10 then 1 to 10^6
n = np.arange(1, 10, 0.1)
f = np.log2(n)
g = np.log10(n)

plt.plot(n,g,"pink", label="g(n)")
plt.plot(n,f,"blue", label="f(n)")
plt.legend()
```

Here, it seems apparent that f(n) grows much faster than g(n), but let's zoom out...

```python
n = np.arange(1, pow(10,6), 0.1)
f = np.log2(n)
g = np.log10(n)

plt.plot(n,g,"pink", label="g(n)")
plt.plot(n,f,"blue", label="f(n)")
plt.legend()
```

Unexpected! This seems to yield a similar result. Certainly g(n) is flattening out, but that is typical log behavior, and it is not flattening out to zero. Maybe we need the limit test to collect all the necessary information. Important to note that the functions are still in significant growth orders relative to each other because it doesn't look like either are "flattening out" relative to each other and going to 0.




```python
#Deliverable 5.2: Limit f(n)/g(n) from 1 to 10 then 1 to 10^6
n = np.arange(1, 10, 0.1)
f = np.log2(n)
g = np.log10(n)
lim = f/g

plt.plot(n,lim,"black", label="lim")
plt.legend()
```

From 1 to 10, the limit seems to be a constant above 0. So, the functions are the same O.O.G. Let's zoom out...

```python
n = np.arange(1, pow(10,6), 0.1)
f = np.log2(n)
g = np.log10(n)
lim = f/g

plt.plot(n,lim,"black", label="lim")
plt.legend()
```

It appears that both reflect the same constant above 0 as the limit of f(n)/ g(n) goes to infinity, so f(n) is in Theta(g(n)), Omega(g(n)), and O(g(n)).
